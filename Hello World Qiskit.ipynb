```python
import qiskit
print(qiskit.__version__)
```

```python
from qiskit import QuantumCircuit

# Créer un circuit avec 2 qubits et 2 bits
qc = QuantumCircuit(2, 2)

# Appliquer une porte Hadamard sur les qubits 0 et 1 : les met en état de superposition
qc.h(0)
qc.h(1)

# Mesurer les qubits 0 et 1 dans les bits classiques 0 et 1
qc.measure([0, 1], [0, 1])

# Dessiner le circuit
qc.draw(output='mpl')
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

# Créer le circuit avec 2 qubits et 2 bits classiques
qc = QuantumCircuit(2, 2)
qc.h(0)
qc.h(1)
qc.measure([0, 1], [0, 1])

# Utiliser le simulateur quantique
simulator = AerSimulator()

# Exécuter 1024 fois
job = simulator.run(qc, shots=1024)
result = job.result()

# Obtenir les comptages
counts = result.get_counts()
print(counts)
```

```python
from qiskit.visualization import plot_histogram

# Afficher l'histogramme
plot_histogram(counts)
```
